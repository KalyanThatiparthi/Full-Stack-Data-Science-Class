# IF Statement

```python
x=3
if x==3:
    print('Value',x,'is found')
```

# IF-ELSE Statement

```python
x=2
if x>0:
    print(x,'is Positive')
else:
    print(x,'is Negative')
```

# IF-ELSE IF Statement 

```python
x=0
if x>0:
    print(x, 'is Positive')
elif x<0:
    print(x, 'is Negative')
else:
    print('Value is Zero')
```

# SWITCH Statement

```python
day = int(input('enter day count'))

match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case _:
        print("Invalid day!")
```

```python
# Constant Expression
x = 10
y = 5

match x + y:
    case 15:
        print("Result is 15.")
    case 20:
        print("Result is 20.")
    case _:
        print("No match found.")
```

```python
# Limited to Certain Types
grade = input('enter grade')

match grade:
    case 'A':
        print("Excellent!")
    case 'B':
        print("Good!")
    case _:
        print("Not specified.")
```

# Ternary Expression Conditional Statement

```python
x = 10

print("x is positive" if x > 0 else "x is not positive")
```

```python
# Nested Ternary Condition
x = 0
print("x is positive" if x > 0 else (
    "x is not positive" if x < 0 else "x is zero"))
```
