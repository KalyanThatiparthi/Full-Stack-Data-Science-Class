# plot() - A versatile command

```python
x1 = np.linspace(0, 10, 50)

# create a plot figure
#fig = plt.figure()

plt.plot(x1, np.sin(x1), ':')
plt.plot(x1, np.cos(x1), '>')
plt.plot(x1, np.tan(x1), '-')
plt.show()
```

# State-machine interface

```python
x=np.linspace(1,21,5)
```

```python
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")
plt.legend()

plt.show()
```

# Formatting the style of plot

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'b:')
plt.axis([0, 6, 0, 20])
plt.show()
```

# Working with NumPy arrays

```python
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

# Object-Oriented API

```python
fig, ax = plt.subplots(2)
# Call plot() method on the appropriate object
ax[1].plot(x1, np.sin(x1), 'b-')
ax[0].plot(x1, np.cos(x1), 'g-');
```

# Objects and Reference

```python
fig = plt.figure()

x2 = np.linspace(0, 5, 10)
y2 = x2 ** 2

axes = fig.add_axes([0.5,0.5,0.8, 0.8])

axes.plot(x2, y2, 'r')

axes.set_xlabel('x2')
axes.set_ylabel('y2')
axes.set_title('title');
```

```python
x4=range(1,6)
plt.plot(x4)
plt.plot(x4, [i*1.5 for i in x4])
plt.plot(x4, [i-2 for i in x4])
plt.plot(x4, [i+3 for i in x4])
plt.plot(x4, [i//4 for i in x4])
```

# Error Bar Chart

```python
x9=np.arange(0,4,0.2)
y9=np.exp(-x9)
e1=0.1 * np.abs(np.random.randn(len(y9)))
plt.errorbar(x9,y9,yerr=e1, fmt='.-')
plt.show()
```

# Stacked Bar Chart

```python
A=[15,30,45,22]
B=[15,25,50,20]
z2=range(4)
plt.bar(z2,A, color='b')
plt.bar(z2,B, color='r',bottom=A)
plt.show()
```

# Pie Chart

```python
plt.figure(figsize=(7,7))
x10=[35,25,20,20]
labels=['Computers','Electronics','Mechanical','Chemical']
plt.pie(x10,labels=labels, autopct=lambda p:f'{int(round(p*sum(x10)/100))}')
#plt.pie(x10,labels=labels, autopct='%1f.1f%%')
plt.show()
```

```python
data3=np.random.randn(10)
plt.boxplot(data3)
plt.show()
```

# Adding a grid

```python
x15=np.arange(1,5)
plt.plot(x15,x15*2,x15,x15/2.0,x15,x15-3,x15,x15+4)
plt.grid(True)
plt.show()
```

# Handling axes

```python
x15=np.arange(1,5)
plt.plot(x15,x15*2,x15,x15/2.0,x15,x15-3,x15,x15+4)
plt.axis()
plt.grid(True)
plt.axis([0,5,-1,13])
plt.show()
```

# Handling X and Y ticks

```python
u=[5,3,7,6,2,1,9,4,8]
plt.plot(u)
plt.xticks([0,2,4,6,8])
plt.yticks([2,4,6,8,10])
plt.show()
```
