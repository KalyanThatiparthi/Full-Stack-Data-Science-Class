# WHILE Statement

```python
i=0
while i<=2:
    print('Hi')
    i+=1
print('Kalyan')
```

```python
i = 1          # initializing

while i<=3:    # condition
    print('data science :', i)
    i = i + 1  # increment
```

```python
# Nested While
i = 0
while i<=3:
    print('Hello',end=' ') # when we mention end then new line will not create
    j = 1
    while j<=3:
        print('Welcome')
        j = j + 1 
    i = i + 1
    print() 
```

```python
i = 4
while i>=1:
    print('FSDS Class') # when we mention end then new line will not create
    j = 2
    while j>=0:
        print(' Loops' )
        j-=1
        
    i -=1
    print()
```

```python
i = 1
while i <= 2 :
    j = 0
    while  j <= 2 :
        print(i+j, end=" ")
        j += 1
    print()
    i += 1
```

```python
i = 3
while i >= 1 :
    j = 3
    while  j >= 0 :
        print(i-j, end=" ")
        j -= 1
    print()
    i -= 1
```
